```python
# Imports
import pandas as pd
import plotly.express as px
import os
import panel as pn
from pathlib import Path
from dotenv import load_dotenv
pn.extension("plotly")
```

```python
# Setting up API credentials
load_dotenv()
map_box_api = os.getenv('MAPBOX_API_KEY')

# Setting the Mapbox API
px.set_mapbox_access_token(map_box_api)
```

```python
# Read in data
crypto_map = pd.read_csv('crypto_map.csv')
major_blockchain_establishments = crypto_map[['id', 'Latitude', 'Longitude', 'Company']].dropna()
major_blockchain_establishments.head()
```

```python
# Plotting the map of Blockchain Tech Establishments
blockchain_map = px.scatter_mapbox(
    major_blockchain_establishments,
    lat='Latitude',
    lon='Longitude',
    color='Company',
    zoom = 0,
    title="World's Leading Companies that Incorporate Blockchain"
)
#blockchain_map.show()
```

```python
map_as_column = pn.Column(
    "## 50 Top Blockchain Companies across the Globe",
    blockchain_map
)

map_as_column
```
